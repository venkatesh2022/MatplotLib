```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
exp_vals=[1400,600,300,410,250]
exp_labels=["Home Rent","Food","Phone/Internet? Bill","Car","Other utilities"]
```

```python
plt.bar(exp_vals)
```

```python
plt.pie(exp_vals,labels=exp_labels)
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.324,autopct="%0.1f%%")
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.324,autopct="%0.1f%%",
        explode=[0,0.6,0.2,0,0],startangle=45)
plt.show()
```

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.324,autopct="%0.1f%%",
        explode=[0,0.6,0.2,0,0],startangle=45)
plt.show()
plt.savefig("piechart.jpg")
```
